```python
#import the three files


import pandas as pd
import numpy as np

df1=pd.read_csv('file1.csv')
df2=pd.read_csv('file2.csv')
df3=pd.read_csv('file3.csv')
```

```python
#replace the % sign in df1 and df2; df3 does not have the % sin

df1['Customer Lifetime Value'] = df1['Customer Lifetime Value'].str.replace('%','')
```

```python
#change the headers to lower cases

df1.columns=df1.columns.str.lower()
df2.columns=df2.columns.str.lower()
df3.columns=df3.columns.str.lower()

#replace the % sign in df1 and df2; df3 does not have the % sin

df2['customer lifetime value'] = df2['customer lifetime value'].str.replace('%','')
```

```python
#replace the "state"-column with an "st"

df3.columns=['customer', 'st', 'customer lifetime value', 'education', 'gender',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'total claim amount', 'vehicle class']
```

```python
#amend the columns of df2 and df3 to the order of df1

df2=df2[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
    
df3=df3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]    
```

```python
#connect the three lists

data=pd.concat([df1,df2,df3],ignore_index=True)
```

```python
data=data.drop(['education','number of open complaints'],axis=1)
```

```python
#define a string as float

data['customer lifetime value']=data['customer lifetime value'].astype(float)
data['customer lifetime value']=100*data['customer lifetime value']
```

```python
#drop duplicates

data=data.drop_duplicates()
```
